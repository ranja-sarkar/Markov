```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#% matplotlib inline
```

```python

filename1 = r'results_w1d_m3hr_nu0.1_gam0.0025.csv'
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d')
df1 = pd.read_csv(filename1, parse_dates = ['timestamp'], date_parser = dateparse)
df1.set_index('timestamp', inplace = True)


filename2 = r'results_w1d_m3hr_nu0.1_gam0.005.csv'
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d')
df2 = pd.read_csv(filename2, parse_dates=['timestamp'], date_parser = dateparse)
df2.set_index('timestamp',inplace = True)


filename3 = r'results_w1d_m3hr_nu0.1_gam0.0075.csv'
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d')
df3 = pd.read_csv(filename3, parse_dates = ['timestamp'], date_parser = dateparse)
df3.set_index('timestamp', inplace = True)

```

```python
gamma1 = df1['flips_ratio'].tolist()
gamma2 = df2['flips_ratio'].tolist()
gamma3 = df3['flips_ratio'].tolist()
```

```python
flips_data = [gamma1, gamma2, gamma3]
```

```python
# Create a figure instance
fig = plt.figure(1, figsize = (9, 6))

# Create an axes instance
ax = fig.add_subplot(111)

# Create the boxplot
bp = ax.boxplot(flips_data, patch_artist = True)

## Custom x-axis labels
ax.set_xticklabels(['gamma = 0.0025', 'gamma = 0.005', 'gamma = 0.0075'])
ax.set_ylabel('Flips Ratio', color = '#000000')
ax.set_title('Flips Ratio (nu = 0.1)')

## change outline color, fill color and linewidth of the boxes
for box in bp['boxes']:
    # change outline color
    box.set( color = '#000000', linewidth = 2)
    # change fill color
    #box.set( facecolor = '#ffffff' )
    box.set( facecolor = '#6a8f61' )
    

## change color and linewidth of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color = '#000000', linewidth = 2)

## change color and linewidth of the caps
for cap in bp['caps']:
    cap.set(color = '#000000', linewidth = 2)

## change color and linewidth of the medians
for median in bp['medians']:
    median.set(color = '#000000', linewidth = 2)

## change the style of fliers and their fill
for flier in bp['fliers']:
    flier.set(marker = 'o', color = '#e7298a', alpha = 0.5)

# Save the figure
#fig.savefig('fig1.png', bbox_inches='tight')
```
