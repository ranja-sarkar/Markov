```python
import pandas as pd
```

```python
df = r'Brunei_Champion_GT5_tst_pred_w1d_m4hr_svm_nu0.01_krnrbf_gam0.01_01Aug2018_01Jul2019.csv'
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
dfoutput = pd.read_csv(df, parse_dates = ['timestamp'], date_parser = dateparse)
dfoutput.set_index(['timestamp'], inplace = True)
dfoutput['date'] = dfoutput.index
```

```python
dfoutput.prediction = dfoutput.prediction.replace(1, 0)
dfoutput.prediction = dfoutput.prediction.replace(-1, 1)

dfoutput['Anomalies'] = 0
dfoutput.loc[(dfoutput['system_status'] == 'online') & (dfoutput['prediction'] == 1), 'Anomalies'] = 1

dfoutput['Alerts'] = 0
dfoutput.loc[(dfoutput['system_status'] == 'online') & (dfoutput['alert'] == 1) & (dfoutput['prediction'] == 1), 'Alerts'] = 1
```

```python
df1 = dfoutput[['Anomalies','Alerts']].resample('W').sum()
```
