```python
#libraries

import pandas as pd
import numpy as np
from numpy import cov

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#load input time-series (data)

df = pd.read_csv('GK.csv', parse_dates = ['timestamp'], low_memory = False)
df.set_index('timestamp', inplace = True)

#df.shape
```

```python
#variables/tags (considered 3 out of 170 here)

df1 = df.columns[0:3]
```

```python
#df2 = pd.DataFrame(df, columns = df1)
#mat = df2.cov()
#mat
```

```python
#time-series plot of tags

%matplotlib inline

plt.style.use('seaborn-darkgrid')
plt.figure(figsize = (12, 9))
plt.xlabel('Timestamp')

for i in range(0, 3):
    plt.plot(df.index, df[df1[i]], linestyle = '-', label = df1[i])
    
plt.legend()     
plt.show()
```

```python
#pairplot distributions of tags

sns.set(style = "ticks", color_codes = True)
sns.pairplot(df[df1], diag_kind = 'kde')
```

```python
#correlations between tags

correlations = df[df1].corr(method = 'spearman')
correlations.style.background_gradient().set_precision(2)
```

```python
#boxplot of a tag

plt.style.use('seaborn-darkgrid')
df2 = pd.DataFrame(df, columns = df1)
df2.boxplot()
```
